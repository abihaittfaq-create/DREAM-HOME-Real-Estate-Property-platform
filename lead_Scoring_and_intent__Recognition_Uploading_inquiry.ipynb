```python
# ================================
# Lead Scoring + Intent Recognition
# Using inquiries.json
# ================================

# Import Libraries
import pandas as pd
import json
from google.colab import files

# Step 1: Upload inquiries JSON file
uploaded = files.upload()

# Automatically get uploaded filename
file_name = list(uploaded.keys())[0]

# Step 2: Load JSON Data
with open(file_name, "r") as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# Check Dataset
print("Dataset Loaded Successfully!")
print(df.head())
print("\nColumns:")
print(df.columns)


# ================================
# Step 3: Intent Recognition
# ================================

def get_intent(message):

    message = str(message).lower()

    if "ready to buy" in message or "buy" in message:
        return "Buying Intent"

    elif "contact me" in message or "call me" in message:
        return "Contact Request"

    elif "need more details" in message or "details" in message or "information" in message:
        return "Information Request"

    elif "exploring options" in message or "explore" in message or "looking" in message:
        return "Browsing"

    else:
        return "Other"


# Apply Intent Recognition
df["Intent"] = df["message"].apply(get_intent)



# ================================
# Step 4: Lead Scoring
# ================================

def lead_score(row):

    score = 0

    # Timeline Score
    timeline = str(row["timeline"]).lower()

    if "1 month" in timeline:
        score += 40

    elif "3 months" in timeline:
        score += 25

    elif "6 months" in timeline:
        score += 15

    else:
        score += 5


    # Intent Score
    if row["Intent"] == "Buying Intent":
        score += 40

    elif row["Intent"] == "Contact Request":
        score += 30

    elif row["Intent"] == "Information Request":
        score += 20

    elif row["Intent"] == "Browsing":
        score += 10


    return score



# Apply Lead Score
df["Lead Score"] = df.apply(lead_score, axis=1)



# ================================
# Step 5: Lead Category
# ================================

def lead_category(score):

    if score >= 70:
        return "Hot Lead"

    elif score >= 40:
        return "Warm Lead"

    else:
        return "Cold Lead"



df["Lead Category"] = df["Lead Score"].apply(lead_category)



# ================================
# Step 6: Show Final Result
# ================================

result = df[
    [
        "inquiryId",
        "message",
        "Intent",
        "timeline",
        "Lead Score",
        "Lead Category"
    ]
]

print("\nFinal Lead Scoring Result:")
display(result)



# ================================
# Step 7: Save Output CSV
# ================================

df.to_csv("Lead_Scoring_Output.csv", index=False)

print("\nDone! File saved as Lead_Scoring_Output.csv")

# Download file
files.download("Lead_Scoring_Output.csv")
# User Input Testing

user_message = input("Enter your inquiry: ")

# Intent Recognition
user_intent = get_intent(user_message)

# User details
user_budget = int(input("Enter your budget: "))
user_timeline = input("Enter your timeline: ")
user_location = input("Enter your location: ")


# Create user data
user_data = {
    "message": user_message,
    "budget": user_budget,
    "timeline": user_timeline,
    "location": user_location,
    "Intent": user_intent
}


# Lead Score
user_score = lead_score(user_data)

# Lead Category
user_category = lead_category(user_score)


# Final Result
print("\n--- Lead Analysis ---")
print("Message:", user_message)
print("Intent:", user_intent)
print("Lead Score:", user_score)
print("Lead Category:", user_category)
```
