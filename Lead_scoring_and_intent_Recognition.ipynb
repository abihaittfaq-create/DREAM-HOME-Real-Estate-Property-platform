```python
from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

labels = [
    "Buy Property",
    "Rent Property",
    "Schedule Visit",
    "Price Inquiry",
    "General Question"
]
```

```python
def lead_score(message):
    score = 0
    msg = message.lower()

    if "buy" in msg:
        score += 40

    if "budget" in msg or "$" in msg:
        score += 20

    if "visit" in msg or "schedule" in msg:
        score += 20

    if "today" in msg or "tomorrow" in msg:
        score += 20

    if score >= 80:
        priority = "High"
    elif score >= 50:
        priority = "Medium"
    else:
        priority = "Low"

    return score, priority
```

```python
message = input("Enter Customer Message: ")

# Intent Recognition
intent = classifier(message, labels)

# Lead Score
score, priority = lead_score(message)

print("\n========== RESULT ==========")
print("Customer Message:", message)
print("Intent:", intent["labels"][0])
print("Lead Score:", score)
print("Priority:", priority)
```
